```python
#This script returns UMLS CUIs based on an input file of strings, where each line in txt file is a separate string.
#If no results are found for a specific string, this will be noted in output and output file.
#Each set of results for a string is separated in the output file with '***'.


import requests
import argparse
api_key = 'c6109e1b-7f0a-463f-9a11-1a5342006c1a'
import requests

# Set API key and other parameters
api_key = 'c6109e1b-7f0a-463f-9a11-1a5342006c1a'
version = 'current'
outputfile = 'testing_output.txt'
inputfile = 'tp201587a.txt'  # Change this to your actual input file
sabs = None  # Example: 'MSH,SNOMEDCT_US,RXNORM'
ttys = None  # Example: 'PT,SY,IN'

base_uri = 'https://uts-ws.nlm.nih.gov'
string_list = []

# Read input file
with open(inputfile, encoding='utf-8') as f:
    for line in f:
        if line.strip():  # Ignore empty lines
            string_list.append(line.strip())

print("Input strings:", string_list)

# Process each string
with open(outputfile, 'w', encoding='utf-8') as o:
    for string in string_list:
        page = 0
        o.write(f'SEARCH STRING: {string}\n\n')
        
        while True:
            page += 1
            path = f'/search/{version}'
            query = {'string': string, 'apiKey': api_key, 'rootSource': sabs, 'termType': ttys, 'pageNumber': page}
            output = requests.get(base_uri + path, params=query)
            output.encoding = 'utf-8'
            
            output_json = output.json()
            results = output_json.get('result', {}).get('results', [])
            
            if not results:
                if page == 1:
                    print(f'No results found for {string}\n')
                    o.write('No results found.\n\n')
                break
            
            for item in results:
                o.write(f"UI: {item.get('ui', 'N/A')}\n")
                o.write(f"Name: {item.get('name', 'N/A')}\n")
                o.write(f"URI: {item.get('uri', 'N/A')}\n")
                o.write(f"Source Vocabulary: {item.get('rootSource', 'N/A')}\n\n")
            
        o.write('***\n\n')

print(f'Results saved to {outputfile}')
```
